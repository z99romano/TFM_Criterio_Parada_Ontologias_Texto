# Experimento 5 — Extracción de relaciones mediante LLM local

En este experimento se evalúa el uso de un modelo de lenguaje local (LLM) para la extracción automática de relaciones semánticas a partir de texto biográfico del corpus DICAT.

A diferencia de los experimentos anteriores, basados principalmente en reglas sintácticas y dependencias gramaticales, este enfoque busca aprovechar la capacidad semántica de los modelos de lenguaje para identificar relaciones más complejas y expresadas mediante estructuras lingüísticas variables.

El objetivo inicial del experimento es comprobar si un LLM local es capaz de:

- identificar sujetos, relaciones y objetos explícitos,
- mantener la formulación textual original de la relación,
- evitar inferencias no presentes en el texto,
- y generar salidas estructuradas en formato JSON.

Para preservar la privacidad de los textos y evitar el uso de servicios externos, se utilizará un modelo ejecutado localmente mediante Ollama.

```python
# Librerías básicas para trabajar con datos, JSON y peticiones HTTP
import pandas as pd
import json
import requests
```

```python
# URL local del servidor de Ollama
OLLAMA_URL = "http://localhost:11434/api/generate"

# Modelo que utilizaremos para la extracción de relaciones
MODEL_NAME = "qwen2.5:7b-instruct"
```

```python
# Función auxiliar para enviar un prompt al modelo local de Ollama
def call_ollama(prompt, model=MODEL_NAME):

    # Datos que se envían al servidor local de Ollama
    payload = {
        "model": model,
        "prompt": prompt,
        "stream": False,
        "temperature": 0
    }

    # Petición HTTP al modelo local
    response = requests.post(OLLAMA_URL, json=payload)

    # Si hay error en la petición, Python lo mostrará aquí
    response.raise_for_status()

    # Devolvemos únicamente el texto generado por el modelo
    return response.json()["response"]
```
